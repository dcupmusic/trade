```python
import vectorbtpro as vbt
import numpy as np
import pandas as pd

vbt.settings.set_theme('dark')
vbt.settings['plotting']['layout']['width'] = 800
vbt.settings['plotting']['layout']['height'] = 400
```

```python
data = vbt.BinanceData.pull(
    "SOLUSDT",
    start="2023-01-01",
    timeframe="1 hour"
)
```

```python
data.plot().show()
```

```python
open_price = data.get('Open')
high_price = data.get('High')
low_price = data.get('Low')
close_price = data.get('Close')
```

```python
# vbt.phelp(vbt.MA.run)
```

```python
adx = vbt.ADX.run(high_price, low_price, close_price, window=14)
rsi = vbt.RSI.run(close_price)
sma = vbt.MA.run(close_price, window=20)
```

```python
data.data['SOLUSDT']['ADX'] = adx.adx
data.data['SOLUSDT']['RSI'] = rsi.rsi
data.data['SOLUSDT']['SMA'] = sma.ma
```

```python
data.data['SOLUSDT']
```

```python
data.data['SOLUSDT']['Return'] = data.data['SOLUSDT']['Close'].pct_change(1).shift(-1)
data.data['SOLUSDT']['target'] = np.where(data.data['SOLUSDT']['Return'] > 0, 1, 0)
# data.data['SOLUSDT']
```

```python
data.data['SOLUSDT'] = data.data['SOLUSDT'].dropna()
predictors_list = ['ADX', 'RSI', 'SMA']
X = data.data['SOLUSDT'][predictors_list]
# print(X)

y = data.data['SOLUSDT']['target']
# y
```

```python
from sklearn.model_selection import train_test_split
split_percentage = 0.7
split = int(split_percentage*len(X))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1-split_percentage)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_leaf=5)
# clf
```

```python
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
report = classification_report(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
# Store the test dataset in new dataframe df_split
df_split = data.data['SOLUSDT'][split:].copy()

# Store the decision tree's predicted output to signal column of df_split dataframe
df_split.loc[:, "signal"] = y_pred

# df_split
```

```python
signal = df_split['signal']
entries = signal == 1
exits = signal == 0
```

```python
pf = vbt.Portfolio.from_signals(
    close=df_split.Close, 
    entries=entries, 
    exits=exits,
    size=100,
    size_type='value',
    init_cash='auto'
)
```
